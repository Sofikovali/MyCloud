# Токен для Яндекс.Метрики

В этом шаге мы пройдем процесс получения/генерации токена для запросов к API Яндекс.Метрики. Далее в Модуле мы будем работать с токеном, который был сгенерирован именно таким образом. 

Токены представляют собой набор символов, который подставляется в каждый запрос к API. Без токена система не отдаст вам данные. Токен может сгенерировать только тот пользователь, который имеет доступ к интерфейсу системы и данным. По сути, токен — это аналог логина и пароля, но в более удобном для использования в коде виде.

Для получения токенов к сервисам Яндекса существует специальная служба https://oauth.yandex.ru/

Здесь вы заводите приложение и уже приложению выдаете права на получение определенных данных. После создания приложения и назначения прав доступа можно сгенерировать токен.

Например, ваше приложение может иметь доступ на чтение отчетов Яндекс.Метрики и Яндекс.Директа. Таким образом удобно контролировать доступы приложений к разным группам данных. 

Приложения для работы с сервисами Яндекса могут иметь разные схемы доступа. В этом модуле мы рассмотрим самый простой вариант работы с API Яндекс.Метрики с помощью так называемого отладочного токена. У сервисов Яндекса есть и другие способы авторизации (например, по протоколу OAuth 2.0, который мы рассмотрим в следующем модуле). Теперь уже видно, что дата лежит во вложенном словаре, а значения метрик заключены в листе под ключом 'metrics' (порядок метрик такой же, как мы указывали в переменной metrics_string). Выведем значения метрик:

Посчитайте среднее значение доли новых посетителей в этом ответе по дням. Т. е. посчитайте отношение суммы новых пользователей за каждый день к сумме всех пользователей за каждый день. Ответ округлите до второго знака после запятой. Формат ответа: 0.79



```python
answer1 = 0
answer2 = 0
for line in data['data']:
    visit_date = line['dimensions'][0]['name'] 
    visits = line['metrics'][0]
    users = line['metrics'][1]
    pageviews = line['metrics'][2]
    percent_new_visitors = line['metrics'][3]
    answer1 += percent_new_visitors/100*users
    answer2 += users
    print(visit_date, visits, users, pageviews, percent_new_visitors)
print (answer1/answer2)
```

В этом блоке мы научимся решать очень распространенную проблему выгрузки больших таблиц.

Это связано с тем, что в любой системе есть ограничения по количеству строк в отчете. API Метрики за один запрос может вернуть максимум 100 тысяч строк. Для больших сайтов есть множество отчетов, которые многократно превосходят этот лимит.

Например, для одного нашего проекта выгрузка отчета по поисковым запросам, страницам входа и дате визита за полгода заняла на диске более 6 гигабайт.

Однако и для небольших сайтов существуют отчеты, когда приходится выгружать частями. Например, когда вам необходимо детально проанализировать посещения сайта по минутам или секундам и разбивке по дополнительной метрике (например, источникам трафика). Такие задачи часто встречаются при анализе оффлайновых рекламных кампаний.

В текущем блоке мы научимся выгружать отчет любого размера, используя параметры limit (необходимое количество строк в ответе, максимум 100 тысяч) и offset (номер первой строки выборки). Этот метод полностью аналогичен тому, как мы выгружали список пользователей ВКонтакте.

Как обычно, задаем параметры выгрузки. В качестве измерения (строк отчета) будет выступать время визита ym:s:dateTime. Из метрик для простоты оставим только визиты:

```python
import requests
from yaml import load
from pprint import pprint
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
headers = {'Authorization': 'OAuth ' + token}

startDate = '2018-02-12'
endDate = '2018-02-18'
counter = '21075004'
metrics = ['ym:s:visits']
dimensions = ['ym:s:dateTime']
```

```python
metrics_string = ','.join(metrics)
dimensions_string = ','.join(dimensions)
params = {
    'date1': startDate,
    'date2': endDate,
    'id': counter,
    'dimensions': dimensions_string,
    'metrics': metrics_string,
}
```

```python
r = requests.get(API_URL, params=params, headers=headers)
data = r.json()
print(data)
```

По умолчанию Метрика возвращает первые 100 строк. Добавим параметры limit и offset к запросу.

Для наглядности выведем первые 5 строк и обернем процесс выгрузки в отдельную функцию:

```python
def report_for_offset_value(offset, n_rows):
    """Выгрузка n_rows строк отчета Метрики для заданного значения offset"""   
    params = {
        'date1': startDate,
        'date2': endDate,
        'id': counter,
        'dimensions': dimensions_string,
        'metrics': metrics_string,
        'offset': offset,
        'limit': n_rows
    }   
    r = requests.get(API_URL, params=params, headers=headers)
    return r.json()
```

```python
offset= 1
n_rows= 5
data = report_for_offset_value(offset, n_rows)
pprint(data)
```

Выведем на экран значения измерений и метрик:

```python
for line in data['data']:
    time_of_visit = line['dimensions'][0]['name']
    visits = line['metrics'][0]  
    print(time_of_visit, visits)
```

В текущей формулировке задачи у нас появляется проблема: мы заранее не знаем, какие параметры будут переданы в запросе API.

Например, сейчас в нашем запросе нет фильтра (параметр filters, который мы использовали в прошлом шаге), но он может понадобиться. Или может понадобиться изменить точность запроса (параметр accuracy). Также в функцию запроса могут передаваться и вспомогательные переменные, не касающиеся Метрики.

Для учета таких ситуаций используется следующий прием: в аргументах функции напишем **kwargs и смотрим, что представляет собой эта переменная:

```python
def make_request(**kwargs):
    print(kwargs)
make_request(date1 = '2018-02-12', date2 = '2018-02-18')
```

Переменная kwargs (кстати, ее название сложилось исторически, можно выбрать любое другое название, главное — ставить две звездочки) преобразуется в словарь с аргументами, которые мы можем использовать для запроса к API Метрики. Если вам нужно передать в функцию набор значений без указания их имен, то можно использовать *args.

С помощью этого приема наша функция запроса к Метрике станет универсальной и сократится до одной строки:

```python
def make_request(**kwargs):
    """Запрос к API Яндекс.Метрики. Возвращает JSON ответа"""
    return requests.get(API_URL, params=kwargs, headers=headers).json()
```

```python
data = make_request(date1=startDate, date2 = endDate, dimensions = ','.join(dimensions), metrics = ','.join(metrics), id = counter,accuracy = 1)
```

```python
pprint(data)
```

Мы получили ответ в JSON-формате, который неплохо бы перевести в наглядную таблицу. Поскольку мы теперь заранее не знаем, как называются столбцы и строки, давайте сделаем универсальный алгоритм, который преобразует данные из JSON-формата в наглядный набор листов.

Рассмотрим сначала лист с измерениями. Выведем первый элемент:

```python
for line in data['data']:
    print(line['dimensions'])
    break
```

Он представляет собой лист из словарей одинаковой структуры. Давайте достанем все значения измерений для каждого ключа 'name':

```python
for line in data['data']:    
    for dimension_value in line['dimensions']:
        print(dimension_value['name'])      
    break
```

Или в более короткой записи:

```python
for line in data['data']:    
    print( [x['name'] for x in line['dimensions']] )        
    break
```

Добавим аналогичный цикл для значений метрик (строчку с выводом измерений пока закомментируем). Значения метрик уже являются листом, поэтому никаких преобразований они не требуют:

```python
for line in data['data']:    
    # print( [x['name'] for x in line['dimensions']] )
    print( line['metrics'] )
       
    break
```

Давайте теперь объединим измерения и метрики каждой строчки отчета в лист report:

```python
report = []
for line in data['data']:
    dimensions_list = [x['name'] for x in line['dimensions']]
    metrics_list = line['metrics']   
    report.append( dimensions_list + metrics_list )   
    print(report)       
    break
```

Оформим наш цикл в функцию и уберем break:

```python
def json_handler(data):
    """Преобразование ответа из JSON в список листов"""
    report = []
    for line in data['data']:
        dimensions_list = [x['name'] for x in line['dimensions']]
        metrics_list = line['metrics']
        report.append( dimensions_list + metrics_list )    
    return report
```

```python
json_handler(data)
```

Такой лист уже гораздо легче обрабатывать и сохранять как выгрузку в файл. Итого у нас получилось две довольно универсальных функции: для выгрузки отчета из API и обработки JSON-ответа в читаемый формат.
