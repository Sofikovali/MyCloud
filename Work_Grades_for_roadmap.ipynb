```python
import pandas as pd
import math
```

```python
data = pd.read_excel('plan.xlsx')
```

```python
data = data[data.User1.notnull()] 
data = data.drop('Topic', 1)
```

```python
data['means'] = 0.0
for line in data.index:
    data['means'][line] = (data['User1'][line]+ data['User2'][line]+data['User3'][line] + data['User4'][line])/4
```

```python
count = 0
for line in data.index:
    s = data['User1'][line] 
    l = data['User2'][line]
    a = data['User3'][line]
    t = data['User4'][line]
    if abs(s-l)>3:
        count += 1
        print('Задача № {} "{}"\nсоперники: Соня,{:.0f} и Игорь,{:.0f}'.format(data['Number'][line],data['Issue'][line] , s,l ))
        print('Остальные: Таня,{:.0f}, Аля,{:.0f}\n'.format(a,t))
        n = int(input())
        data['means'][line] = n
    elif abs(l-a)>3:
        count += 1
        print('Задача № {} "{}"\nCоперники: Игорь,{:.0f} и Алевтина,{:.0f}'.format(data['Number'][line],data['Issue'][line] , l,a ))
        print('Остальные: Таня,{:.0f}, Соня,{:.0f}\n'.format(t,s))
        n = int(input())
        data['means'][line] = n
    elif abs(s-a)>3:
        count += 1
        print('Задача № {} "{}"\nCоперники: Соня,{:.0f} и Алевтина, {:.0f}'.format(data['Number'][line],data['Issue'][line] , s,a ))
        print('Остальные: Таня,{:.0f},Игорь,{:.0f}\n'.format(t,l))
        n = int(input())
        data['means'][line] = n
    elif abs(t-a)>3:
        count +=1
        print('Задача № {} "{}"\nCоперники: Таня,{:.0f} и Алевтина, {:.0f}'.format(data['Number'][line],data['Issue'][line] , t,a ))
        print('Остальные: Соня,{:.0f},Игорь,{:.0f}\n'.format(s,l))
        n = int(input())
        data['means'][line] = n
    elif abs(t-s)>3:
        count += 1
        print('Задача № {} "{}"\nCоперники: Таня,{:.0f} и Соня, {:.0f}'.format(data['Number'][line],data['Issue'][line] , t,s ))
        print('Остальные: Аля,{:.0f}, Игорь,{:.0f}\n'.format(a,l))
        n = int(input())
        data['means'][line] = n
    elif abs(t-l)>3:
        count+= 1
        print('Задача № {} "{}"\nCоперники: Таня,{:.0f} и Игорь, {:.0f}'.format(data['Number'][line],data['Issue'][line] , t,l ))
        print('Остальные: Соня,{:.0f}, Аля,{:.0f}\n'.format(s,a))
        n = int(input())
        data['means'][line] = n
print('Всего спорных задач: {}'.format(count))
```
