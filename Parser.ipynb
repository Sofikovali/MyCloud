```python
import pandas as pd
import codecs
categories = []
with codecs.open('Categories.txt','r', "utf_8_sig") as f:
    for line in f:
        category =line.strip()#убираем служебный символ переноса строки, который дает пробел
        print(category)
        categories.append(category)
```

```python
df=pd.DataFrame(categories,columns=['col1'])  
```

```python
df.head()
```

```python
df = pd.DataFrame(df.col1.str.split('/').tolist(),
                                   columns = ['RU','EN'])
```
