```python
def json_handler(data):
    """Преобразование ответа из JSON в список листов"""
    report = []
    for line in data['data']:
        dimensions_list = [x['name'] for x in line['dimensions']]
        metrics_list = line['metrics']
        report.append( dimensions_list + metrics_list )    
    return report
```

```python
frame1 = pd.DataFrame(report, columns=['date','deviceCategory', 'mobilePhone', 'mobilePhoneModel','users'])
```

```python
frame2 = pd.read_csv('arpu_revenue_data.csv', sep= '\t', decimal=',')
```

```python
frame2['revenue'] = pd.to_numeric(frame2['revenue'])
frame2
```

```python
frame1a = frame1.fillna('None')
```

```python
report_df_sum = frame1a.groupby(['deviceCategory', 'mobilePhone', 'mobilePhoneModel'])['users'].sum().reset_index()
```

```python
report1 = report_df_sum[report_df_sum['users']>= 3]
```

```python
pivot = report1.merge(frame2, on=['deviceCategory', 'mobilePhone', 'mobilePhoneModel'], how = 'left')
```

```python
pivot['arpu'] = pivot.apply(lambda x: float(x['revenue'])/x['users'], axis = 1)
```

```python
result = pivot[pivot['deviceCategory'] =='ПК']
```
